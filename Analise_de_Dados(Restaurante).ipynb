```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
```

```python
g = sns.FacetGrid(tips, hue="sex")  # col="time", row="sex", height=3.5, aspect= 3, col_wrap=3 (para col="size"),
g.map(sns.scatterplot, 'total_bill', 'tip')
g.add_legend()
g.set_axis_labels("Total Bill($)", "Tip($)")
g.fig.suptitle("Gráfico de ScatterPlot- - Total Bill por Sexo", fontsize=10)
g.fig.subplots_adjust(top=0.9)


variance_by_sex = tips.groupby('sex')['total_bill'].var()
print("Variância de total_bill por sexo:")
for sex, variance in variance_by_sex.items():
    print(f"{sex}: {variance:.2f}")
print("----------------------------------------")

covariance = tips['total_bill'].cov(tips['tip'])
print("Covariância entre total_bill e tip:")
print(f"{covariance:.2f}")

print("----------------------------------------------------")

std_dev_total_bill = tips['total_bill'].std()
print("Desvio padrão de total_bill:")
print(f"{std_dev_total_bill:.2f}")

print("------------------------------------------------------")
```

```python
sns.violinplot(tips, x='smoker', y='total_bill', hue="sex", fill=False)
plt.title('Gráfico de Violino - Total Bill por Sexo')
plt.xlabel('Smoker')
plt.ylabel('Total Bill')

variance_by_category = tips.groupby(['sex', 'smoker'])['total_bill'].var()
print("Variância de total_bill por sexo e fumante:")
for (sex, smoker), variance in variance_by_category.items():
    print(f"{sex}, {smoker}: {variance:.2f}")

print("----------------------------------------")

covariance = tips['total_bill'].cov(tips['tip'])
print("Covariância entre total_bill e tip:")
print(f"{covariance:.2f}")

print("----------------------------------------------------")

std_dev_total_bill = tips['total_bill'].std()
print("Desvio padrão de total_bill:")
print(f"{std_dev_total_bill:.2f}")

print("------------------------------------------------------")

plt.show()

# Outros comandos importantes:
# 'scale': (Este é usado para ajustar a largura dos violinos.)
# 'inner': Este é usado para controlar a representação dos dados observados no interior dos violinos.
# 'bw': Este é usado para controlar a largura de banda do KDE no interior do violino.
```

```python
sns.catplot(data=tips, x='smoker', y='total_bill', hue="sex", kind="box")
plt.title('Gráfico de Fumantes - Total Bill por Sexo e Hábito de Fumar')
plt.xlabel('Fumante')
plt.ylabel('Total Bill')
plt.show()
```

```python
g = sns.relplot(
    data=tips, x="total_bill", y="tip", col="time",
    hue="time", style="sex",
    palette=["b", "r"]
)

g.fig.subplots_adjust(top=0.9)
plt.subplots_adjust(wspace=0.1)
g.fig.suptitle('Relação entre conta total e gorjeta')

plt.show()

variance_total_bill_by_sex = tips.groupby('sex')['total_bill'].var()
print("Variância de total_bill por sexo:")
for sex, variance in variance_total_bill_by_sex.items():
    print(f"{sex}: {variance:.2f}")

variance_tip_by_sex = tips.groupby('sex')['tip'].var()
print("\nVariância de gorjeta por sexo:")
for sex, variance in variance_tip_by_sex.items():
    print(f"{sex}: {variance:.2f}")

variance_total_bill_by_time = tips.groupby('time')['total_bill'].var()
print("\nVariância de total_bill por time:")
for time, variance in variance_total_bill_by_time.items():
    print(f"{time}: {variance:.2f}")

variance_tip_by_time = tips.groupby('time')['tip'].var()
print("\nVariância de tip por time:")
for time, variance in variance_tip_by_time.items():
    print(f"{time}: {variance:.2f}")
print("------------------------------------------------")

cov_total_bill_tip = tips[['total_bill', 'tip']].cov().iloc[0, 1]
print(f'Covariância entre conta total e gorjeta: ')
print(f"{cov_total_bill_tip:.2f}")
print("------------------------------------------------")

std_total_bill = tips['total_bill'].std()
print(f'Desvio padrão da conta total:')
print(f"{std_total_bill:.2f}")
print("------------------------------------------------")

std_tip = tips['tip'].std()
print(f'Desvio padrão da gorjeta:')
print(f"{std_tip:.2f}")
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(15, 6))

# Exemplo 1
sns.histplot(data=tips, x="tip", hue="sex", element="step", bins=20, ax=axs[0])
axs[0].set_title('Distribuição das Gorjetas por Sexo')
axs[0].set_xlabel('tip')
axs[0].set_ylabel('Count')


# Exempo 2
sns.histplot(data=tips, x="day", hue="sex", multiple="dodge", shrink=.8, ax=axs[1])
axs[1].set_title('Contagem de registros por sexo para cada dia da semana')

plt.subplots_adjust(wspace=0.5)
plt.show()
```
